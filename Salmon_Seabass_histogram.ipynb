# Fish Classification Using Histogram Analysis

In this program, we aim to categorize fish based on their length and brightness using histograms. We start by creating a function that minimizes error by placing a threshold on the histogram. Then, we apply the golden section search method to find the optimal threshold.

### Step-by-Step Process:

1. **Thresholding for Error Minimization**:
    - First, a function is implemented to minimize the error by placing a threshold on the histogram. The objective is to find the threshold that divides the data in a way that minimizes the error.
  
2. **Golden Section Search**:
    - The golden section search method is used to find the optimal threshold. This approach is chosen for its efficiency in finding a local minimum in one-dimensional problems, such as determining the best threshold.

3. **Feature Comparison**:
    - We compare the two features, **length** and **brightness**, to determine which one is more effective for classification. The comparison shows that **brightness** is a more effective feature for categorization.

4. **Creating a Dataset**:
    - To establish a correlation between the histogram data, we start creating a dataset for the fish based on the mean and variance of each histogram. This dataset helps in understanding the distribution of the data.

5. **Classification with Linear Classifier**:
    - Using the generated dataset, we proceed to classification. A **linear classifier** is applied using the **Powell optimization method** for efficient function minimization. Additionally, the **Nelder-Mead** method, a second-order optimization technique, is used for fine-tuning the classification model.

6. **Final Classification**:
    - By applying these optimization techniques, we are able to categorize the fish based on their brightness and length features, utilizing the optimal thresholds and the linear classifier to achieve accurate results.

### Conclusion:
By using histograms to categorize fish based on their features and applying optimization techniques, we can effectively classify the fish with the highest accuracy. Brightness proves to be a more effective feature than length in this classification task.

```python
import math

def golden_section_search_minimize_error(calculate_error_func, a, b, tolerance=1e-3):


    inv_phi = (math.sqrt(5) - 1) / 2
    
    inv_phi_squared = (3 - math.sqrt(5)) / 2

    if a > b:
        a, b = b, a

    x1 = a + inv_phi_squared * (b - a)
    x2 = a + inv_phi * (b - a)

    f_x1, _, _ = calculate_error_func(x1, None)
    
    f_x2, _, _ = calculate_error_func(x2, None)


    while (b - a) > tolerance:
        if f_x1 < f_x2:
            b = x2
            x2 = x1
            f_x2 = f_x1
            x1 = a + inv_phi_squared * (b - a)
            f_x1, _, _ = calculate_error_func(x1, None)
        else:
            a = x1
            x1 = x2
            f_x1 = f_x2
            x2 = a + inv_phi * (b - a)
            f_x2, _, _ = calculate_error_func(x2, None)

    return (a + b) / 2
```

```python
import math

def golden_section_search_minimize_error(calculate_error_func, a, b, length_data, tolerance=1e-3):
    inv_phi = (math.sqrt(5) - 1) / 2
    inv_phi_squared = (3 - math.sqrt(5)) / 2

    if a > b:
        a, b = b, a

    x1 = a + inv_phi_squared * (b - a)
    x2 = a + inv_phi * (b - a)

    # تبدیل xها به نزدیک‌ترین عدد صحیح چون threshold باید صحیح باشد
    x1 = round(x1)
    x2 = round(x2)

    f_x1, _, _ = calculate_error_func(x1, length_data)
    f_x2, _, _ = calculate_error_func(x2, length_data)

    while (b - a) > tolerance:
        if f_x1 < f_x2:
            b = x2
            x2 = x1
            f_x2 = f_x1
            x1 = a + inv_phi_squared * (b - a)
            x1 = round(x1)
            f_x1, _, _ = calculate_error_func(x1, length_data)
        else:
            a = x1
            x1 = x2
            f_x1 = f_x2
            x2 = a + inv_phi * (b - a)
            x2 = round(x2)
            f_x2, _, _ = calculate_error_func(x2, length_data)

    return round((a + b) / 2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from sklearn.model_selection import train_test_split

length_data = np.array([
    [2, 7, 4], [3, 13, 4], [4, 16, 5], [5, 12, 1], [6, 19, 7],
    [7, 20, 5], [8, 18, 4], [9, 18, 7], [10, 17, 12], [11, 9, 10],
    [12, 8, 14], [13, 8, 7], [14, 12, 12], [15, 10, 14], [16, 8, 15],
    [17, 7, 14], [18, 4, 18], [19, 6, 19], [20, 8, 9], [21, 7, 5],
    [22, 6, 4], [23, 0, 0]
])

brightness_data = np.array([
    [0.4, 0., 0.], [0.8, 0., 2.], [1.2, 0., 8.], [1.6, 0., 4.],
    [2., 0., 4.], [2.4, 0., 7.], [2.8, 0., 10.], [3.2, 0., 12.],
    [3.6, 1., 8.], [4., 1., 6.], [4.4, 0., 6.], [4.8, 3., 8.],
    [5.2, 3., 4.], [5.6, 4., 3.], [6., 6., 2.], [6.4, 6., 1.],
    [6.8, 5., 1.], [7.2, 12., 0.], [7.6, 9., 0.], [8., 9., 0.],
    [8.4, 3., 0.], [8.8, 2., 0.], [9.2, 1., 0.], [9.6, 1., 0.],
    [10., 2., 0.]
])


def calculate_mean_variance(data, column):
    """محاسبه میانگین و واریانس برای داده‌های هیستوگرامی"""
    values = []
    for row in data:
        value = row[0]
        count = int(row[column])
        values.extend([value] * count)
    values = np.array(values)
    mean = np.mean(values) if len(values) > 0 else 0
    variance = np.var(values) if len(values) > 0 else 1e-6
    return mean, variance

mean_length_free, var_length_free = calculate_mean_variance(length_data, 1)
mean_length_salmon, var_length_salmon = calculate_mean_variance(length_data, 2)
mean_brightness_free, var_brightness_free = calculate_mean_variance(brightness_data, 1)
mean_brightness_salmon, var_brightness_salmon = calculate_mean_variance(brightness_data, 2)

def generate_paired_data(mean_length, var_length, mean_brightness, var_brightness, n_samples=150, correlation=0.3):
    """تولید داده‌های جفتی با توزیع نرمال دو متغیره و خروجی به صورت لیست"""
    mean = [mean_length, mean_brightness]
    cov = [
        [var_length, correlation * np.sqrt(var_length * var_brightness)],
        [correlation * np.sqrt(var_length * var_brightness), var_brightness]
    ]
    data = np.random.multivariate_normal(mean, cov, n_samples)
    data[:, 0] = np.clip(data[:, 0], 2, 28)
    data[:, 1] = np.clip(data[:, 1], 0.4, 9.6)
    data_list = [[float(length), float(brightness)] for length, brightness in data]
    return data_list


np.random.seed(0)
free_data = generate_paired_data(mean_length_free, var_length_free, mean_brightness_free, var_brightness_free, 300)
salmon_data = generate_paired_data(mean_length_salmon, var_length_salmon, mean_brightness_salmon, var_brightness_salmon, 300)

paired_data_list = []
for sample in free_data:
    paired_data_list.append([sample[0], sample[1], 0])
for sample in salmon_data:
    paired_data_list.append([sample[0], sample[1], 1])

X = np.array([[sample[0], sample[1]] for sample in paired_data_list])
y = np.array([sample[2] for sample in paired_data_list])

X_min = np.min(X, axis=0)
X_max = np.max(X, axis=0)
X_scaled = (X - X_min) / (X_max - X_min)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

y_train_signed = np.where(y_train == 1, 1, -1)
y_test_signed = np.where(y_test == 1, 1, -1)

def linear_error(params):
    a, b, c = params
    outputs = a * X_train[:, 0] + b * X_train[:, 1] + c
    loss = np.log(1 + np.exp(-y_train_signed * outputs))
    return np.sum(loss)

initial_guess = [0, 0, 0]
res_linear = minimize(linear_error, initial_guess, method='Powell')
a_opt_linear, b_opt_linear, c_opt_linear = res_linear.x

outputs_linear_train = a_opt_linear * X_train[:, 0] + b_opt_linear * X_train[:, 1] + c_opt_linear
preds_linear_train = (outputs_linear_train > 0).astype(int)
error_linear_train = np.sum(preds_linear_train != y_train)
total_samples_train = len(y_train)
accuracy_linear_train = 1 - error_linear_train / total_samples_train


outputs_linear_test = a_opt_linear * X_test[:, 0] + b_opt_linear * X_test[:, 1] + c_opt_linear
preds_linear_test = (outputs_linear_test > 0).astype(int)
error_linear_test = np.sum(preds_linear_test != y_test)
total_samples_test = len(y_test)
accuracy_linear_test = 1 - error_linear_test / total_samples_test

print("کلاسیفایر خطی:")
print(f"بهترین ضرایب: a = {a_opt_linear:.4f}, b = {b_opt_linear:.4f}, c = {c_opt_linear:.4f}")
print(f"داده‌های آموزش - تعداد خطاها: {int(error_linear_train)}، دقت: {accuracy_linear_train:.4f}")
print(f"داده‌های آزمایش - تعداد خطاها: {int(error_linear_test)}، دقت: {accuracy_linear_test:.4f}")

def quadratic_error(params):
    a, b, c, d, e, f = params
    outputs = (a * X_train[:, 0]**2 + b * X_train[:, 1]**2 + 
               c * X_train[:, 0] * X_train[:, 1] + 
               d * X_train[:, 0] + e * X_train[:, 1] + f)
    loss = np.log(1 + np.exp(-y_train_signed * outputs))
    return np.sum(loss)

initial_guess_quad = [1, -1, 0, 0, 0, 0]
res_quad = minimize(quadratic_error, initial_guess_quad, method='Nelder-Mead')
a_opt_quad, b_opt_quad, c_opt_quad, d_opt_quad, e_opt_quad, f_opt_quad = res_quad.x

outputs_quad_train = (a_opt_quad * X_train[:, 0]**2 + b_opt_quad * X_train[:, 1]**2 + 
                      c_opt_quad * X_train[:, 0] * X_train[:, 1] + 
                      d_opt_quad * X_train[:, 0] + e_opt_quad * X_train[:, 1] + f_opt_quad)
preds_quad_train = (outputs_quad_train > 0).astype(int)
error_quad_train = np.sum(preds_quad_train != y_train)
accuracy_quad_train = 1 - error_quad_train / total_samples_train

outputs_quad_test = (a_opt_quad * X_test[:, 0]**2 + b_opt_quad * X_test[:, 1]**2 + 
                     c_opt_quad * X_test[:, 0] * X_test[:, 1] + 
                     d_opt_quad * X_test[:, 0] + e_opt_quad * X_test[:, 1] + f_opt_quad)
preds_quad_test = (outputs_quad_test > 0).astype(int)
error_quad_test = np.sum(preds_quad_test != y_test)
accuracy_quad_test = 1 - error_quad_test / total_samples_test

print("\nکلاسیفایر درجه دوم:")
print(f"بهترین ضرایب: a = {a_opt_quad:.4f}, b = {b_opt_quad:.4f}, c = {c_opt_quad:.4f}, "
      f"d = {d_opt_quad:.4f}, e = {e_opt_quad:.4f}, f = {f_opt_quad:.4f}")
print(f"داده‌های آموزش - تعداد خطاها: {int(error_quad_train)}، دقت: {accuracy_quad_train:.4f}")
print(f"داده‌های آزمایش - تعداد خطاها: {int(error_quad_test)}، دقت: {accuracy_quad_test:.4f}")

plt.figure(figsize=(10, 6))
plt.scatter(X[y == 0, 0], X[y == 0, 1], color='blue', label='Sea bass (Free)', alpha=0.6)
plt.scatter(X[y == 1, 0], X[y == 1, 1], color='orange', label='Salmon', alpha=0.6)

plt.ylim(0, 10)

x1_vals = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100)
x1_scaled = (x1_vals - X_min[0]) / (X_max[0] - X_min[0])
x2_scaled_linear = -(a_opt_linear * x1_scaled + c_opt_linear) / b_opt_linear
x2_vals_linear = np.clip(x2_scaled_linear * (X_max[1] - X_min[1]) + X_min[1], 0.4, 9.6)
plt.plot(x1_vals, x2_vals_linear, 'r--', label='Linear Decision Boundary')

x2_vals_quad = []
for x1_s in x1_scaled:
    A = b_opt_quad
    B = e_opt_quad + c_opt_quad * x1_s
    C = a_opt_quad * x1_s**2 + d_opt_quad * x1_s + f_opt_quad
    discriminant = B**2 - 4 * A * C
    if discriminant >= 0 and A != 0:
        x2_s1 = (-B + np.sqrt(discriminant)) / (2 * A)
        x2_s2 = (-B - np.sqrt(discriminant)) / (2 * A)
        x2_1 = x2_s1 * (X_max[1] - X_min[1]) + X_min[1]
        x2_2 = x2_s2 * (X_max[1] - X_min[1]) + X_min[1]
        if 0.4 <= x2_1 <= 9.6:
            x2_vals_quad.append(x2_1)
        elif 0.4 <= x2_2 <= 9.6:
            x2_vals_quad.append(x2_2)
        else:
            x2_vals_quad.append(np.nan)
    else:
        x2_vals_quad.append(np.nan)

plt.plot(x1_vals, x2_vals_quad, 'g--', label='Quadratic Decision Boundary')

plt.xlabel('Fish Length')
plt.ylabel('Fish Brightness')
plt.title('Classifiers with Powell Method (Length + Brightness)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(X[y == 0, 0], bins=20, alpha=0.5, color='blue', label='Sea bass (Generated)')
plt.hist(X[y == 1, 0], bins=20, alpha=0.5, color='orange', label='Salmon (Generated)')
plt.title('Length Histogram (Generated Data)')
plt.xlabel('Length')
plt.ylabel('Count')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(X[y == 0, 1], bins=20, alpha=0.5, color='blue', label='Sea bass (Generated)')
plt.hist(X[y == 1, 1], bins=20, alpha=0.5, color='orange', label='Salmon (Generated)')
plt.title('Brightness Histogram (Generated Data)')
plt.xlabel('Brightness')
plt.ylabel('Count')
plt.legend()

plt.tight_layout()
plt.show()
```
